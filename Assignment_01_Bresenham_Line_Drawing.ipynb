1. Write a program to implement Bresenham's line drawing algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_line(line_co_ordinate, X_LIM=0, Y_LIM=0):
    """
    Display line using co-ordinate
    Argument:
        line_co_ordinate (list)     : list of cordinates. shape(no-of-point, 2).
        X_LIM (scaler)              :
        Y_LIM (scaler)              :
    Return:
        none
    """
    fig = plt.figure(figsize = (12, 10))
    co_ordinate = np.array(line_co_ordinate)
    co_ordinate = co_ordinate.T
    x_co_ordinate = co_ordinate[0, :]
    y_co_ordinate = co_ordinate[1, :]

    plt.xlim(0, X_LIM)
    plt.ylim(0, Y_LIM)
    plt.scatter(x_co_ordinate, y_co_ordinate)
    plt.show()
```

```python
def bresenham_line(P, Q):
    """
    Draw line using bresenham line drawing algorithm
    Argument:
        P (tuple)   : starting co-ordinate of line(x1, y1).
        Q (tuple)   : ending co-ordinate of line(x2, y2).
    Return:
        line_co_ordinate (list) :  list of cordinates. shape(no-of-point, 2).

    """
    line_co_ordinate = []
    x1, y1 = P
    x2, y2 = Q
    
    dx = x2 - x1
    dy = y2 - y1

    pk = 2 * dy - dx

    x = x1
    y = y1

    X_LIM = x2 + 5
    Y_LIM = y2 + 5

    while(x <= x2):
        line_co_ordinate.append([x, y])
        if pk < 0:
            pk += 2 * dy
            
        else:
            pk += 2 * dy - 2 * dx
            y = y + 1
        x += 1
    display_line(line_co_ordinate, X_LIM, Y_LIM)
    return line_co_ordinate

```

```python
P = (10, 10)
Q = (50, 50)

points = bresenham_line(P, Q)
print(points)
```
