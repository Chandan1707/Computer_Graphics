```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_obj(ax, obj, c, label):
    """
    print object 
    Argument:
        ax (object)     : graph object contain graph position
        obj (ndarray)   : shape(m, n), object.
        c (string)      : contain which colour object print
        label (string)  : label of object.
    return:
        
    """

    flag = 1
    for i in range(4):
        for j in range(1, 5):
            ax.plot([obj[0, i], obj[0, (i+j)%4]], \
                    [ obj[1, i],  obj[1, (i+j)%4]], \
                    [ obj[2, i], obj[2, (i+j)%4]], c=c, label= label if flag==1 else None)
            flag = 0
```

```python
def translation_3d(obj, t_v, X_LIM=10, Y_LIM=10, Z_LIM=10):
    """
    Translate 3D object
    Argument:
        t_v (list(3, ))     : Translation vector contain t_x, t_y, t_z
        obj (ndarray)       : shape(m, n), it is the object.
        X_LIM (scaler)      : limit of graph along x-axis.
        Y_LIM (scaler)      : limit of graph along y-axis.
        Z_LIM (scaler)      : limit of graph along z-axis.
    Return:
        
    """

    t_x, t_y, t_z = t_v
    fig = plt.figure(figsize = (12, 12))
    ax = plt.axes(projection ='3d')

    # set limit along x, y, z axis
    ax.set_xlim(-X_LIM ,X_LIM)
    ax.set_ylim(-Y_LIM, Y_LIM)
    ax.set_zlim(-Z_LIM, Z_LIM)

    # print x, y, z axis
    ax.plot([-X_LIM, X_LIM], [0, 0], [0, 0], c="r")
    ax.plot([0, 0], [-Y_LIM, Y_LIM], [0, 0], c="r")
    ax.plot([0, 0], [0, 0], [-Z_LIM, Z_LIM], c="r")

    # Translation vector 
    T_v = np.array([[1, 0, 0, t_x], 
                    [0, 1, 0, t_y], 
                    [0, 0, 1, t_z], 
                    [0, 0, 0,   1]])
    
    print_obj(ax, obj, "g", "Original Object")
    obj_prime = np.dot(T_v, obj)
    print_obj(ax, obj_prime, "b", "After Translation")

    ax.legend()
    ax.view_init(elev=20., azim=-35)

    plt.show()
```

```python
# 3D object A(1, 1, 2), B(3, 3, 2), C(1, 5, 2) and D(2, 3, 4)
obj = np.array([[1, 3, 1, 2], 
                [1, 3, 5, 3], 
                [2, 2, 2, 4], 
                [1, 1, 1, 1]])
# Translation vector (t_v = 3i + 3j + 3k)
t_v = [3, 3, 3]

translation_3d(obj, t_v)
```
