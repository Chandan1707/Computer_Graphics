2. Write a program to implement mid-point circle drawing algorithm.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def display_circle(line_co_ordinate, X_LIM=0, Y_LIM=0):
    """
    Display circle using co-ordinate
    Argument:
        line_co_ordinate (list)     : list of cordinates. shape(no-of-point, 2).
        X_LIM (scaler)              :
        Y_LIM (scaler)              :
    Return:
        none
    """
    fig = plt.figure(figsize = (10, 10))
    co_ordinate = np.array(line_co_ordinate)
    co_ordinate = co_ordinate.T
    x_co_ordinate = co_ordinate[0, :]
    y_co_ordinate = co_ordinate[1, :]

    plt.plot([X_LIM, -X_LIM], [0, 0], c="r")
    plt.plot([0, 0], [Y_LIM, -Y_LIM], c="r")
    plt.xlim(-X_LIM, X_LIM)
    plt.ylim(-Y_LIM, Y_LIM)
    plt.scatter(x_co_ordinate, y_co_ordinate)
    plt.show()
```

```python
def mid_point_circle(radious):
    """
    Mid-point circle drawing algorithm.
    Argument:
        radious (scaler) : radious of the circle.
    Return:
        co_ordinates (list) : co-ordinates of the circle.
    
    """
    co_ordinates = []
    x , y = 0, radious

    d = 2 * x - y + 5 / 4

    while(x <= y):
        co_ordinates.append([ x,  y])
        co_ordinates.append([ y,  x])
        co_ordinates.append([ y, -x])
        co_ordinates.append([ x, -y])
        co_ordinates.append([-x, -y])
        co_ordinates.append([-y, -x])
        co_ordinates.append([-y,  x])
        co_ordinates.append([-x,  y])
        if d < 0:
            d = d + 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1

        x += 1
    display_circle(co_ordinates, radious*1.2, radious*1.2)
    return co_ordinates
```

```python
points = mid_point_circle(10)
print(points)
```
